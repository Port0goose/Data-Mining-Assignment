```python
# Unique value counts for all object columns
cat_summary = (df_customer.select_dtypes(include='object').nunique().sort_values(ascending=False).to_frame(name='Unique_Values'))

print("Categorical Column Cardinality:")
display(cat_summary)

for col in df_customer.select_dtypes(include='object').columns:
    uniques = df_customer[col].dropna().unique()
    print(f"\n--- {col} ---")
    print(f"{len(uniques)} unique values")
    print(uniques[:10])
```

### Check for duplicate customers via their Loyalty ID and duplicate flights via their Loyalty ID and date

```python
# Count duplicate customer IDs
dups_customers = df_customer.duplicated(subset=['Loyalty#']).sum()
print(f"Duplicate Loyalty# in CustomerDB: {dups_customers}")

# Inspect any duplicates if found
if dups_customers > 0:
    display(df_customer[df_customer.duplicated(subset=['Loyalty#'], keep=False)])
```

```python
df_customer = df_customer.drop_duplicates(subset=['Loyalty#'], keep='first')
```

```python
# Check for duplicate monthly flight entries
dups_flights = df_flights.duplicated(subset=['Loyalty#','Year','Month']).sum()
print(f"Duplicate (Loyalty#, Year, Month) rows in FlightsDB: {dups_flights}")

# Inspect any duplicates found
if dups_flights > 0:
    display(df_flights[df_flights.duplicated(subset=['Loyalty#','Year','Month'], keep=False)])
```

```python
df_flights = df_flights.drop_duplicates(subset=['Loyalty#','Year','Month'], keep='first')
```

### Check for invalid number ranges

### We'll use inner join to combine both dataframes, since that will only retain the customers with valid flight activity and loyalty records

```python
# Inner join both dataframes together
df_inner_join = pd.merge(df_customer, df_flights, on='Loyalty#', how='inner')
df_inner_join.head()
```

```python
df_inner_join.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Plot value counts for each column
for column in df_inner_join.columns:
    plt.figure(figsize=(12, 6))
    
    # For categorical/object columns
    if df_inner_join[column].dtype == 'object' or df_inner_join[column].nunique() < 50:
        value_counts = df_inner_join[column].value_counts().head(20)
        
        # Bar plot
        plt.subplot(1, 2, 1)
        value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title(f'{column} - Top 20 Values (Bar Chart)')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        
        # Pie chart for top 10
        plt.subplot(1, 2, 2)
        value_counts.head(10).plot(kind='pie', autopct='%1.1f%%', startangle=90)
        plt.title(f'{column} - Top 10 Values (Pie Chart)')
        plt.ylabel('')
        
    # For numerical columns
    else:
        plt.subplot(1, 2, 1)
        df_inner_join[column].hist(bins=30, color='lightcoral', edgecolor='black')
        plt.title(f'{column} - Distribution (Histogram)')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        plt.subplot(1, 2, 2)
        df_inner_join[column].plot(kind='box', color='lightgreen')
        plt.title(f'{column} - Box Plot')
        plt.ylabel(column)
    
    plt.tight_layout()
    plt.show()
```
